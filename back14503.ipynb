```python
space_size = list(map(int,input().split(" ")))
robot_info = list(map(int,input().split(" ")))
space = []
for i in range(space_size[0]):
    a = list(map(int,input().split(" ")))
    space.append(a)
robot_location = robot_info[0:2]
robot_direction = robot_info[2]

class robot():
    def __init__(self):
        self.robot_location= robot_location         #list [r,c]
        self.robot_direction= robot_direction       # 0 N/ 1 W/ 2 S/ 3 E
        self.space = space
        self.space[self.robot_location[0]][self.robot_location[1]] = 2
        self.clear_num = 1
        self.close = 0
    def clean(self):
        while True:
            if self.close != 4:
                if self.robot_direction == 0:
                    if self.space[self.robot_location[0]][self.robot_location[1]-1] == (1 or 2):
                        self.robot_direction = 1
                        self.close += 1
                    elif self.space[self.robot_location[0]][self.robot_location[1]-1] == 0:
                        self.clear_num += 1
                        self.robot_location[1] -= 1
                        self.space[self.robot_location[0]][self.robot_location[1]] = 2
                        self.close = 0
                elif self.robot_direction ==1:
                    if self.space[self.robot_location[0]+1][self.robot_location[1]] == (1 or 2):
                        self.robot_direction = 2
                        self.close += 1
                    elif self.space[self.robot_location[0]+1][self.robot_location[1]] == 0:
                        self.clear_num += 1
                        self.robot_location[0] += 1
                        self.space[self.robot_location[0]][self.robot_location[1]] = 2
                        self.close = 0
                elif self.robot_direction ==2:
                    if self.space[self.robot_location[0]][self.robot_location[1]+1] == (1 or 2):
                        self.robot_direction = 3
                        self.close += 1
                    elif self.space[self.robot_location[0]][self.robot_location[1]+1] == 0:
                        self.clear_num += 1
                        self.robot_location[1] += 1
                        self.space[self.robot_location[0]][self.robot_location[1]] = 2
                        self.close = 0
                else:
                    if self.space[self.robot_location[0]-1][self.robot_location[1]] == (1 or 2):
                        self.robot_direction = 0
                        self.close += 1
                    elif self.space[self.robot_location[0]-1][self.robot_location[1]] == 0:
                        self.clear_num += 1
                        self.robot_location[0] -= 1
                        self.space[self.robot_location[0]][self.robot_location[1]] = 2
                        self.close = 0
            elif self.close == 4:
                if self.robot_direction == 0:
                    if self.space[self.robot_location[0]+1][self.robot_location[1]] == 1:
                        break
                    else:
                        self.robot_location[0] += 1
                        self.close = 0
                elif self.robot_direction == 1:
                    if self.space[self.robot_location[0]][self.robot_location[1]+1] == 1:
                        break
                    else:
                        self.robot_location[1] += 1
                        self.close = 0
                elif self.robot_direction == 2:
                    if self.space[self.robot_location[0]-1][self.robot_location[1]] == 1:
                        break
                    else:
                        self.robot_location[0] -= 1
                        self.close = 0
                else:
                    if self.space[self.robot_location[0]][self.robot_location[1]-1] == 1:
                        break
                    else:
                        self.robot_location[1] -= 1
                        self.close = 0
        return self.clear_num

robott = robot()
numm = robott.clean()
print(numm)
                
```
